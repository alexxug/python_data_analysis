# Analysing the Stroop Effect

In a Stroop task, participants are presented with a list of words, with each word displayed in a color of ink. The participant’s task is to say out loud the color of the ink in which the word is printed. The task has two conditions: a congruent words condition, and an incongruent words condition. 

In the congruent words condition, the words being displayed are color words whose names match the colors in which they are printed: for example RED, BLUE. 

In the incongruent words condition, the words displayed are color words whose names do not match the colors in which they are printed: for example PURPLE, ORANGE. In each case, we measure the time it takes to name the ink colors in equally-sized lists. Each participant will go through and record a time from each condition.

Task can be performed using following link: 
https://faculty.washington.edu/chudler/java/timestc.html

A hypothesis regarding the outcome of the task is created, then the task is done by myself. Then some data collected from others who have performed the same task is analysed.Finally results were interpreted in terms of the hypothesis.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("stroopdata.csv")
df.head(2)
```

** Identifying independent variable and dependent variable **

Independent Variable - Task type (congrunt or incongruent)

Dependent Variable - Time taken to finish the task in seconds

** Deciding an appropriate set of hypotheses for this task, statistical tests that need to be performed. **

μc = population mean completion time for task with congrunt words. μi = population mean completion time for task with incongruent words.

Null hypothesis(Ho): Poulation mean completion times for both congrunt words task and incongruent words task are the same. (μc=μi)
Alternative hypothesis(Ha): Population mean completion times for two tasks are not equal. In other words there is a difference between those. (μc ≠ μi) (μc - μi ≠ 0)

In this experiment same user group is tested under two dependent scenarios. Further none of the population realted statistical values are provided. So that a paired t-test is most suitable because it can comapre the means of two related groups to find out whether there is a difference between the means, without much information about population data.

** Conducting some descriptive statistics regarding this dataset. **
